**Part 1: Analyzing Best-Seller Trends Across Product Categories**

Objective: Understand the relationship between product categories and their best-seller status.

1. Crosstab Analysis:

- Create a crosstab between the product category and the isBestSeller status.
- Are there categories where being a best-seller is more prevalent?
Hint: one option is to calculate the proportion of best-sellers for each category and then sort the categories based on this proportion in descending order.

2. Statistical Tests:
- Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
- Compute Cramér's V to understand the strength of association between best-seller status and category.

3. Visualizations:
- Visualize the relationship between product categories and the best-seller status using a stacked bar chart.


```python
# 1a) Create a crosstab between the product category and the isBestSeller status:
crosstab_result = pd.crosstab(df['category'], df['isBestSeller'])
crosstab_result
```

```python
crosstab_result["proportion"]= crosstab_result[True]/(crosstab_result[False]+crosstab_result[True])*100
crosstab_result
```

```python
# 1b) Are there categories where being a best-seller is more prevalent?
# Hint: one option is to calculate the proportion of best-sellers for each category and then sort the categories based on this proportion in descending order.
(crosstab_result/crosstab_result.sum()).sort_values(by = True, ascending = False).round(2)
```

Answer 1b) Being a bestseller is most prevalent in the categories Grocery and Health & Personal Care. 

```python
# 2a) Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
from scipy.stats import chi2_contingency

# Chi-square test for 'MSZoning' and 'SaleCondition'
chi2_statistic, chi2_p_value, _, _ = chi2_contingency(crosstab_result)

chi2_statistic, chi2_p_value
# (36540.20270061387, 0.0)

# direct interpretation of the value of 0 - assuming that this assumption is true, 
# => the prob of observing our data is basically zero! chi 2 value refers to probability of obersing the given data
# below conclusion from that interpretation
```

Answer 2a) The null hypothesis of the Chi-square test is that the two categorical variables are independent. Given the p-value of 0 (!), we reject the null hypothesis. This implies that there's a significant association between best-seller distribution and product category in the dataset.

```python
# 2b) Compute Cramér's V to understand the strength of association between best-seller status and category.
from scipy.stats.contingency import association

# Computing the association between variables in 'crosstab_result' using the "cramer" method!
association(crosstab_result, method="cramer")
```

Answer 2b) The Cramér's V value of 0.122 suggests a weak association between the two variables. Though statistically significant, the strength of this relationship is not very strong in practical terms.

```python
# 3) Visualize the relationship between product categories and the best-seller status using a stacked bar chart.
crosstab_result.plot(kind="bar", stacked=True)
plt.show()
```

```python
# Graph above is overcrowded, limiting result to top 10 categories according to bestseller status 
top10 = (crosstab_result.sort_values(by = True, ascending = False).round(2).head(10))
top10.plot(kind="bar", stacked=True)
plt.show()
```

**Part 2: Exploring Product Prices and Ratings Across Categories and Brands**

**Objective: Investigate how different product categories influence product prices.**

**0. Preliminary Step: Remove outliers in product prices.**

For this purpose, we can use the IQR (Interquartile Range) method. Products priced below the first quartile minus 1.5 times the IQR or above the third quartile plus 1.5 times the IQR will be considered outliers and removed from the dataset. The next steps will be done with the dataframe without outliers.
Hint: you can check the last Check For Understanding at the end of the lesson EDA Bivariate Analysis for a hint on how to do this.

**1. Violin Plots:**
- Use a violin plot to visualize the distribution of price across different product categories. Filter out the top 20 categories based on count for better visualization.
- Which product category tends to have the highest median price? Don't filter here by top categories.

**2. Bar Charts:**
- Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
- Which product category commands the highest average price? Don't filter here by top categories.

**3. Box Plots:**
- Visualize the distribution of product ratings based on their category using side-by-side box plots. Filter out the top 10 categories based on count for better visualization.
- Which category tends to receive the highest median rating from customers? Don't filter here by top categories.


```python
# 0. Preliminary Step: Remove outliers in product prices.

def tukeys_test_outliers(data): 
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    # Define bounds for the outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify the outliers
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    
    return outliers
```

```python
data_series = df['price']  # replace 'ColumnName' with your specific column
outliers = tukeys_test_outliers(data_series)
print(outliers)
# 327.688 out of 2.443.651 are outliers! #only for normALLY DISTRIBUTED 
```

```python
# Remove outliers below 1.5 and above 3.5 IQR
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR = Q3 - Q1
    
# Define bounds for the outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
    
# Identify the outliers
clean_df = df[(df["price"] > lower_bound) & (df["price"] < upper_bound)]
clean_df # use this df starting from now! 2115963 (2.443.651-327.688), so looks right 
```

```python
# P2/1. Violin Plots: Use a violin plot to visualize the distribution of price across different product categories. 
sns.violinplot(data=clean_df, x='category', y='price')
plt.show()
```

```python
# Filter out the top 20 categories based on count for better visualization. 

# Get the top 20 categories by count
top20_cat = clean_df['category'].value_counts().nlargest(20).index
# Filter the dataframe
filtered_df = clean_df[clean_df['category'].isin(top20_cat)]
# Plot the violin plot
sns.violinplot(data=filtered_df, x='category', y='price', palette="coolwarm")
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.show()
```

```python
# Which product category tends to have the highest median price? Don't filter here by top categories.
top_median_prices = clean_df.groupby('category')['price'].median().sort_values(ascending=False)
print(top_median_prices)
```

Answer 1a) Desktop PCs are the category with the highest median price of 74.00 $. 

```python
# Part II / 2. Bar Charts: # Create a bar chart comparing the average price of products for the top 10 product categories (based on count).

# Get the top 10 categories by count
top10_cat = clean_df['category'].value_counts().nlargest(10).index
# Filter the dataframe
top10_df = clean_df[clean_df['category'].isin(top10_cat)]
sns.barplot(data=top10_df, x='category', y='price', palette="coolwarm")
plt.xticks(rotation=90) 
plt.show()
```

```python
# Which product category commands the highest average price? Don't filter here by top categories.
top_median_prices = clean_df.groupby('category')['price'].mean().sort_values(ascending=False)
print(top_median_prices)
# Answer: Its Motherboards with a price of 68.772432 $
```

```python
# Part II / 3. Box Plots:
# Visualize the distribution of product ratings based on their category using side-by-side box plots. 
# Filter out the top 10 categories based on count for better visualization.
top10_cat = clean_df['category'].value_counts().nlargest(10).index
# Filter the dataframe
top10_df = clean_df[clean_df['category'].isin(top10_cat)] 
sns.boxplot(data=top10_df, x='category', y='stars', palette="coolwarm")
plt.xticks(rotation=90) 
plt.show()
```

```python
# Now evrything without the 0 stars ratings 
top10_cat = clean_df['category'].value_counts().nlargest(10).index

# Filter the dataframe to include only top 10 categories and exclude 0-star ratings
top10_df = clean_df[(clean_df['category'].isin(top10_cat)) & (clean_df['stars'] > 0)]

# Create the box plot
sns.boxplot(data=top10_df, x='category', y='stars', palette="coolwarm")
plt.xticks(rotation=90) 
plt.show()
```

```python
# Which category tends to receive the highest median rating from customers? Don't filter here by top categories.
top_median_rating = clean_df.groupby('category')['stars'].median().sort_values(ascending=False)
print(top_median_rating)
# Computer Memory with a rating of 4.7
```
